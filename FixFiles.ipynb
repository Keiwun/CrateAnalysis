```python
from MuDataFrame import *
import copy


def getCombinedMDFO(runStart, runEnd):
    r1, r2 = runStart, runEnd
    lead_files = [
        "processed_data/run{}.csv".format(i) for i in range(r1, r2 + 1)
    ]

    mdfo_c = []  # collection of objects

    for file in lead_files:
        mdfo_c.append(MuDataFrame(file))  # Muon Data Frame Object for Lead

    mdf_list = [i.events_df for i in mdfo_c]
    mdf_lead = mdfo_c[0].getMergedMDF(mdf_list)

    mdfo_lead = copy.copy(mdfo_c[0])
    mdfo_lead.events_df = mdf_lead

    mdfo_lead.longDataMode()

    mdf_lead = mdfo_lead.events_df
    mdfo_lead.og_df = mdf_lead.copy()
    return mdfo_lead


def mergeMDFO(mdfos):
    mdf_list = [i.events_df for i in mdfos]
    mdf_lead = mdfo_c[0].getMergedMDF(mdf_list)

    mdfo_lead = copy.copy(mdfo_c[0])
    mdfo_lead.events_df = mdf_lead

    mdfo_lead.longDataMode()

    mdf_lead = mdfo_lead.events_df
    mdfo_lead.og_df = mdf_lead.copy()
    return mdfo_lead


def getEmissionOutputCSV(df, fileName):
    df.events_df.reset_index(drop=True, inplace=True)
    df.events_df.to_csv(fileName)


def getMergedMDF(df_list):
    df = pd.concat(df_list).reset_index()
    return df
```

```python
mdfo_calib = getCombinedMDFO(439, 445)
```

# Combine

## Fix time and event num

```python
last_time = mdfo_calib_og.get("event_time")[-1]
last_event = mdfo_calib_og.get("event_num")[-1] + 1


mdfo_calib.events_df["event_time"] = mdfo_calib.events_df["event_time"].values + last_time
mdfo_calib.events_df["event_num"] = mdfo_calib.events_df["event_num"].values + last_event
```

## Join the MDF's

```python
combined_mdf = pd.concat([mdfo_calib_og.events_df, mdfo_calib.events_df], ignore_index=True)
```
